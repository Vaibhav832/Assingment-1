Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. Scatter plot - used to visualize the relationship between two variables. It helps to determine whether there is a correlation between the variables or not.
2. Line plot - used to display the trend of a variable over time. It helps to identify patterns or trends in the data.
3. Bar plot - used to compare different categories of data. It helps to visualize the differences in values between the categories.
4. Histogram - used to show the distribution of a variable. It helps to identify the shape of the data and any outliers.
5. Pairplot - pair plot is a powerful tool for identifying patterns and trends in a dataset, especially when dealing with high-dimensional datasets where it is difficult to visualize all the variables at once.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.<br>
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

fmri=sns.load_dataset('fmri') # Load the dataset fmri
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='region') # Plot the lineplot
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.<br>
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import pandas as pd

try:
    titanic_data = sns.load_dataset("titanic")
except:
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
    titanic_data = pd.read_csv(url)

sns.boxplot(data=titanic_data, x="pclass", y="age")
sns.boxplot(data=titanic_data, x="pclass", y="fare")
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")
# Plot a histogram of prices grouped by cut
sns.histplot(data=diamonds_data, x="price", hue="cut")
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns

iris=sns.load_dataset('iris') # Load the dataset iris
sns.pairplot(iris,hue='species') # Plot the pairplot
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

flights=sns.load_dataset('flights') # Load the dataset flights
# Reshape the data to a matrix
flights_matrix = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_matrix) # Plot the heatmap
```
