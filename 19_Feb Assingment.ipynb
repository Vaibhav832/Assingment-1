Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a Python web framework that allows you to build web applications quickly and easily. It is lightweight and flexible, making it a popular choice for developers who want to get their applications up and running quickly.

Advantages of using Flask Framework:

1. Lightweight: Flask is a lightweight framework, meaning that it has a small core and does not require particular tools or libraries to function. This makes it easy to set up and use, and ensures that it doesn't get in the way of your application's performance.

2. Flexible: Flask is a flexible framework, which means that it can be used for a wide range of applications, from simple microservices to complex web applications. It also allows developers to easily extend and customize its functionality to meet their specific needs.

3. Easy to learn: Flask is relatively easy to learn compared to other web frameworks, which means that developers can quickly get started building applications without spending a lot of time learning the ins and outs of the framework.

4. Large community: Flask has a large and active community of developers who contribute to its development and offer support and guidance to other developers. This makes it easy to find answers to questions and to get help with any issues that may arise.

5. Good documentation: Flask has excellent documentation that is easy to follow and understand. This documentation covers everything from getting started with Flask to advanced topics such as extensions and deployment.

Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

```python
from flask import Flask

app=Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, App Routing refers to the process of mapping URLs to their respective functions or views within a Flask application. App Routing allows developers to define the behavior of their Flask application based on the URL that is being requested by the client.

The primary purpose of App Routing in Flask is to provide a way to handle incoming HTTP requests and map them to specific functions that will generate an appropriate response. This allows developers to create dynamic web applications that can respond to user input and provide customized content based on the user's actions.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask

app=Flask(__name__)

@app.route('/welcome')
def welcome():
    return "Welcome to ABC Corporation"

@app.route('/')
def company_details():
    return "Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

In Flask, the url_for() function is used for URL building. The url_for() function generates a URL based on the name of the function that handles the desired endpoint, and any arguments that the endpoint requires.

Here is an example code snippet that demonstrates how to use the url_for() function in Flask:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the homepage'

@app.route('/user/<username>')
def user_profile(username):
    return f'This is the profile page for {username}'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('user_profile', username='Vaibhav'))

if __name__ == '__main__':
    app.run(host="0.0.0.0", port=5000)
```
